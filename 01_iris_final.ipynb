# Iris Data Analysis

The Iris dataset is a well-known dataset to learn
data analysis. It consists of measurement samples of different types of iris flowers. This exercise is based on an Iris dataset analysis from [Kaggle](https://www.kaggle.com/code/suneelpatel/learn-ml-from-scratch-with-iris-dataset).

The first step is to import the Python libraries.

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_palette('husl')
import matplotlib.pyplot as plt
%matplotlib inline
```

Then read the CSV file and retrieve the data size:

```python
#Load Data
iris = pd.read_csv('iris.csv')
```

```python
#How big is the dataset?
iris.shape
```

Explore the data by retrieving the first and last lines.

```python
# See the first 5 lines
iris.head()
```

```python
# See the last 5 lines
iris.tail()
```

Check that the data is evenly distributed among the three iris
species.

```python
iris['species'].value_counts()
```

Generate a histogram of the petal width of the three iris species. The histogram divides into two histograms, it also shows three peaks, which suggests that the three iris species vary in petal width.

```python
# @title petal_width

from matplotlib import pyplot as plt
iris['petal_width'].plot(kind='hist', bins=20, title='petal_width')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

The petal length histogram divides into two histograms, suggesting that one iris species has a significantly smaller petal length.

```python
# @title petal_length

from matplotlib import pyplot as plt
iris['petal_length'].plot(kind='hist', bins=20, title='petal_length')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

The sepal width histogram suggests that sepal width is not a strong differentiator of the iris species.

```python
# @title sepal_width

from matplotlib import pyplot as plt
iris['sepal_width'].plot(kind='hist', bins=20, title='sepal_width')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

The sepal length histogram is scattered, suggesting that sepal length varies greatly and that it might be a weak iris species differentiator.

```python
# @title sepal_length

from matplotlib import pyplot as plt
iris['sepal_length'].plot(kind='hist', bins=20, title='sepal_length')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

Do statistical analysis of the data and see how the data of the entire sample compares with each of the iris species.

```python
#Analize the data
iris.describe()
```

### Setosa species analysis


```python
iris[iris['species'] == 'setosa'].describe()
```

### Versicolor species analysis

```python
iris[iris['species'] == 'versicolor'].describe()
```

### Iris species analysis

```python
iris[iris['species'] == 'virginica'].describe()
```

As the histograms suggest, the main differentiators of the three Iris species is the petal length and the petal width. The Setosa species petals are significantly smaller than the Virginica and Versicolor. The setosa species also shows less variation in size. The sepal size results to be also a differentiator of the different species, but it's very similar between the Versicolor and the Virginica species.

```python
g = sns.violinplot(y='species', x='sepal_length', data=iris, inner='quartile')
plt.show()
g = sns.violinplot(y='species', x='sepal_width', data=iris, inner='quartile')
plt.show()
g = sns.violinplot(y='species', x='petal_length', data=iris, inner='quartile')
plt.show()
g = sns.violinplot(y='species', x='petal_width', data=iris, inner='quartile')
plt.show()
```

Finally, these graphs show how the values are grouped by species. Sepal length and sepal width show more grouped while petal length and petal width show more differences. Note how the Setosa species patterns are different from the rest of the species.

```python
#iris1 = iris.drop('index', axis=1)
g = sns.pairplot(iris, hue='species', markers='+')
plt.show()
```
